```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("cameraman.tif")
cv2_imshow(img)
```

```python
cv2.resize(img,(100,50))
```

```python
ig = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
cv2_imshow(ig[1])
```

```python
img = 255-ig[1]
cv2_imshow(img)
```

```python
cv2.resize(img,None,fx=0.5,fy=0.5,interpolation= cv2.INTER_AREA)
```
